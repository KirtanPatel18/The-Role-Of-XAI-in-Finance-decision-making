```python
# Cell 5 — Logistic Regression baseline: fit, evaluate (val + test), save artifacts & plots
import os, json
import numpy as np
import joblib
import matplotlib.pyplot as plt
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    classification_report, confusion_matrix, roc_auc_score,
    roc_curve, precision_recall_curve, precision_score, recall_score, f1_score
)
from sklearn.metrics import ConfusionMatrixDisplay

ROOT = '/content/drive/MyDrive/XAI_TRY_1_PROJECT'
ARTIFACTS = os.path.join(ROOT, 'artifacts')
VIZ = os.path.join(ROOT, 'viz')
os.makedirs(ARTIFACTS, exist_ok=True)
os.makedirs(VIZ, exist_ok=True)

# config
RANDOM_STATE = 42

# Build pipeline: use the preprocessor fitted earlier
lr_pipeline = Pipeline([
    ('preproc', preprocessor),  # preprocessor fitted in Cell 4
    ('clf', LogisticRegression(max_iter=1000, class_weight='balanced', random_state=RANDOM_STATE, solver='liblinear'))
])

# Fit on train
print("Fitting Logistic Regression on X_train...")
lr_pipeline.fit(X_train, y_train)
print("LR fitted.")

# Save pipeline
lr_path = os.path.join(ARTIFACTS, 'lr_pipeline.joblib')
joblib.dump(lr_pipeline, lr_path)
print("Saved LR pipeline to:", lr_path)

# Evaluate function (computes metrics, plots, and saves them)
def evaluate_and_save(pipe, X_eval, y_eval, tag):
    # predictions & probs
    y_pred = pipe.predict(X_eval)
    if hasattr(pipe, "predict_proba"):
        y_proba = pipe.predict_proba(X_eval)[:,1]
    else:
        # fallback to decision_function if available
        try:
            y_proba = pipe.decision_function(X_eval)
            # convert to 0-1 range with logistic (not perfect but OK for plotting)
            from scipy.special import expit
            y_proba = expit(y_proba)
        except Exception:
            y_proba = np.zeros(len(y_pred))

    # metrics
    report = classification_report(y_eval, y_pred, output_dict=True)
    cm = confusion_matrix(y_eval, y_pred)
    roc_auc = roc_auc_score(y_eval, y_proba) if len(np.unique(y_eval))>1 else None

    # save metrics JSON
    metrics_out = {
        'classification_report': report,
        'confusion_matrix': cm.tolist(),
        'roc_auc': float(roc_auc) if roc_auc is not None else None
    }
    with open(os.path.join(ARTIFACTS, f'metrics_lr_{tag}.json'), 'w') as f:
        json.dump(metrics_out, f, indent=2)
    print(f"Saved metrics_lr_{tag}.json")

    # save summary CSV (per-class precision/recall/f1)
    rows = []
    for cls, vals in report.items():
        if cls in ['accuracy','macro avg','weighted avg']:
            continue
        row = {'class': cls}
        row.update(vals)
        rows.append(row)
    import pandas as pd
    pd.DataFrame(rows).to_csv(os.path.join(ARTIFACTS, f'model_performance_summary_lr_{tag}.csv'), index=False)
    print(f"Saved model_performance_summary_lr_{tag}.csv")

    # ROC plot
    if roc_auc is not None:
        fpr, tpr, _ = roc_curve(y_eval, y_proba)
        plt.figure(figsize=(6,5))
        plt.plot(fpr, tpr, label=f'ROC (AUC={roc_auc:.3f})')
        plt.plot([0,1],[0,1],'--', color='gray')
        plt.xlabel('False Positive Rate'); plt.ylabel('True Positive Rate')
        plt.title(f'ROC Curve — LR ({tag})')
        plt.legend(loc='lower right')
        plt.grid(True)
        plt.savefig(os.path.join(VIZ, f'roc_lr_{tag}.png'), bbox_inches='tight')
        plt.close()
        print(f"Saved viz/roc_lr_{tag}.png")

    # Precision-Recall plot
    prec, rec, _ = precision_recall_curve(y_eval, y_proba)
    ap = np.trapz(prec, rec)  # approximate area under PR (not average_precision_score but fine for plot label)
    plt.figure(figsize=(6,5))
    plt.plot(rec, prec, label=f'PR (area={ap:.3f})')
    plt.xlabel('Recall'); plt.ylabel('Precision')
    plt.title(f'Precision-Recall — LR ({tag})')
    plt.grid(True)
    plt.savefig(os.path.join(VIZ, f'pr_lr_{tag}.png'), bbox_inches='tight')
    plt.close()
    print(f"Saved viz/pr_lr_{tag}.png")

    # Confusion matrix (absolute and normalized)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['No Default','Default'])
    fig, ax = plt.subplots(figsize=(5,4))
    disp.plot(ax=ax, cmap='Blues', values_format='d')
    ax.set_title(f'Confusion matrix (LR) — {tag}')
    fig.savefig(os.path.join(VIZ, f'confusion_matrix_lr_{tag}.png'), bbox_inches='tight')
    plt.close(fig)
    print(f"Saved viz/confusion_matrix_lr_{tag}.png")

    # Return basic numbers for quick inspection
    return {'roc_auc': roc_auc, 'precision': precision_score(y_eval, y_pred, zero_division=0),
            'recall': recall_score(y_eval, y_pred, zero_division=0), 'f1': f1_score(y_eval, y_pred, zero_division=0)}

# Evaluate on validation set
metrics_val = evaluate_and_save(lr_pipeline, X_val, y_val, 'val')
print("Validation metrics (LR):", metrics_val)

# Evaluate on test set
metrics_test = evaluate_and_save(lr_pipeline, X_test, y_test, 'test')
print("Test metrics (LR):", metrics_test)

print("\nCell 5 complete. LR baseline trained, evaluated, and artifacts + plots saved.")
```

```python
# Cell 6 — Random Forest baseline: fit, evaluate (val + test), save artifacts & plots
import os, json
import numpy as np
import joblib
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (
    classification_report, confusion_matrix, roc_auc_score,
    roc_curve, precision_recall_curve, precision_score, recall_score, f1_score
)
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.calibration import calibration_curve

ROOT = '/content/drive/MyDrive/XAI_TRY_1_PROJECT'
ARTIFACTS = os.path.join(ROOT, 'artifacts')
VIZ = os.path.join(ROOT, 'viz')
os.makedirs(ARTIFACTS, exist_ok=True)
os.makedirs(VIZ, exist_ok=True)

RANDOM_STATE = 42

# Build RF pipeline (use the preprocessor fitted earlier)
rf_pipeline = Pipeline([
    ('preproc', preprocessor),
    ('clf', RandomForestClassifier(n_estimators=200, class_weight='balanced', n_jobs=-1, random_state=RANDOM_STATE))
])

# Fit on train
print("Fitting Random Forest on X_train...")
rf_pipeline.fit(X_train, y_train)
print("RF fitted.")

# Save pipeline
rf_path = os.path.join(ARTIFACTS, 'rf_pipeline.joblib')
joblib.dump(rf_pipeline, rf_path)
print("Saved RF pipeline to:", rf_path)

# evaluation helper (same structure as LR cell)
def evaluate_and_save_rf(pipe, X_eval, y_eval, tag):
    y_pred = pipe.predict(X_eval)
    y_proba = pipe.predict_proba(X_eval)[:,1] if hasattr(pipe, "predict_proba") else np.zeros(len(y_pred))

    report = classification_report(y_eval, y_pred, output_dict=True)
    cm = confusion_matrix(y_eval, y_pred)
    roc_auc = roc_auc_score(y_eval, y_proba) if len(np.unique(y_eval))>1 else None

    metrics_out = {
        'classification_report': report,
        'confusion_matrix': cm.tolist(),
        'roc_auc': float(roc_auc) if roc_auc is not None else None
    }
    with open(os.path.join(ARTIFACTS, f'metrics_rf_{tag}.json'), 'w') as f:
        json.dump(metrics_out, f, indent=2)
    print(f"Saved metrics_rf_{tag}.json")

    # save summary CSV
    rows = []
    for cls, vals in report.items():
        if cls in ['accuracy','macro avg','weighted avg']:
            continue
        row = {'class': cls}
        row.update(vals)
        rows.append(row)
    import pandas as pd
    pd.DataFrame(rows).to_csv(os.path.join(ARTIFACTS, f'model_performance_summary_rf_{tag}.csv'), index=False)
    print(f"Saved model_performance_summary_rf_{tag}.csv")

    # ROC plot
    if roc_auc is not None:
        fpr, tpr, _ = roc_curve(y_eval, y_proba)
        plt.figure(figsize=(6,5))
        plt.plot(fpr, tpr, label=f'ROC (AUC={roc_auc:.3f})')
        plt.plot([0,1],[0,1],'--', color='gray')
        plt.xlabel('False Positive Rate'); plt.ylabel('True Positive Rate')
        plt.title(f'ROC Curve — RF ({tag})')
        plt.legend(loc='lower right')
        plt.grid(True)
        plt.savefig(os.path.join(VIZ, f'roc_rf_{tag}.png'), bbox_inches='tight')
        plt.close()
        print(f"Saved viz/roc_rf_{tag}.png")

    # PR plot
    prec, rec, _ = precision_recall_curve(y_eval, y_proba)
    ap = np.trapz(prec, rec)
    plt.figure(figsize=(6,5))
    plt.plot(rec, prec, label=f'PR area={ap:.3f}')
    plt.xlabel('Recall'); plt.ylabel('Precision'); plt.title(f'Precision-Recall — RF ({tag})')
    plt.grid(True)
    plt.savefig(os.path.join(VIZ, f'pr_rf_{tag}.png'), bbox_inches='tight')
    plt.close()
    print(f"Saved viz/pr_rf_{tag}.png")

    # Confusion matrix (absolute + normalized)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['No Default','Default'])
    fig, ax = plt.subplots(figsize=(5,4))
    disp.plot(ax=ax, cmap='Blues', values_format='d')
    ax.set_title(f'Confusion matrix (RF) — {tag}')
    fig.savefig(os.path.join(VIZ, f'confusion_matrix_rf_{tag}.png'), bbox_inches='tight')
    plt.close(fig)
    print(f"Saved viz/confusion_matrix_rf_{tag}.png")

    # Calibration plot (reliability curve)
    prob_true, prob_pred = calibration_curve(y_eval, y_proba, n_bins=10)
    plt.figure(figsize=(6,5))
    plt.plot(prob_pred, prob_true, marker='o', label='RF')
    plt.plot([0,1],[0,1],'--', color='gray')
    plt.xlabel('Predicted probability'); plt.ylabel('True probability'); plt.title(f'Calibration curve — RF ({tag})')
    plt.grid(True)
    plt.savefig(os.path.join(VIZ, f'calibration_rf_{tag}.png'), bbox_inches='tight')
    plt.close()
    print(f"Saved viz/calibration_rf_{tag}.png")

    return {'roc_auc': roc_auc, 'precision': precision_score(y_eval, y_pred, zero_division=0),
            'recall': recall_score(y_eval, y_pred, zero_division=0), 'f1': f1_score(y_eval, y_pred, zero_division=0)}

# Evaluate on validation
metrics_val_rf = evaluate_and_save_rf(rf_pipeline, X_val, y_val, 'val')
print("Validation metrics (RF):", metrics_val_rf)

# Evaluate on test
metrics_test_rf = evaluate_and_save_rf(rf_pipeline, X_test, y_test, 'test')
print("Test metrics (RF):", metrics_test_rf)

print("\nCell 6 complete. RF baseline trained, evaluated, and artifacts + plots saved.")
```

```python
# SHAP dependence — robust recompute & plotting (new cell)
import os, numpy as np, pickle
import pandas as pd
import matplotlib.pyplot as plt

ROOT = '/content/drive/MyDrive/XAI_TRY_1_PROJECT'
ARTIFACTS = os.path.join(ROOT, 'artifacts')
VIZ = os.path.join(ROOT, 'viz')
os.makedirs(VIZ, exist_ok=True)

# --- 1) Ensure we have a fresh sample and transformed matrix from the SAME preprocessor ---
SHAP_SAMPLE = min(2000, X_train.shape[0])
X_shap_sample = X_train.sample(n=SHAP_SAMPLE, random_state=42)
X_shap_trans = preprocessor.transform(X_shap_sample)  # fitted preprocessor from Cell 4

# --- 2) Robustly build transformed feature names (handles OHE and empty cat lists) ---
def build_feature_names_from_ct(ct):
    names = []
    for name, transformer, cols in ct.transformers_:
        if name == 'remainder':
            continue
        if name == 'num':
            # numeric cols
            names.extend(list(cols))
        elif name == 'cat':
            # categorical pipeline possibly with OneHotEncoder
            try:
                ohe = transformer.named_steps.get('ohe', None)
                if ohe is not None:
                    # sklearn >=1.2: get_feature_names_out available
                    ohe_names = list(ohe.get_feature_names_out(cols))
                    names.extend(ohe_names)
                else:
                    names.extend(list(cols))
            except Exception:
                # fallback: keep raw cat column names (safer than crash)
                names.extend(list(cols))
        else:
            # unknown transformer: attempt to add cols as-is
            try:
                names.extend(list(cols))
            except Exception:
                pass
    # final fallback: if nothing collected, return generic names
    if not names:
        # try to form from X_shap_sample columns
        return [f"feature_{i}" for i in range(X_shap_trans.shape[1])]
    return names

feat_names = build_feature_names_from_ct(preprocessor)

# If lengths still mismatch, adapt names to match the transformed matrix shape
n_cols_trans = X_shap_trans.shape[1]
if len(feat_names) != n_cols_trans:
    # best effort: if feat_names longer, trim; if shorter, append generic suffixes
    if len(feat_names) > n_cols_trans:
        feat_names = feat_names[:n_cols_trans]
    else:
        # extend with generic names
        need = n_cols_trans - len(feat_names)
        feat_names = feat_names + [f"feature_extra_{i}" for i in range(need)]

# --- 3) Load shap_values (from memory or artifacts) ---
try:
    shap_values  # exists in memory
except NameError:
    npz = os.path.join(ARTIFACTS, 'shap_values_sample.npz')
    if not os.path.exists(npz):
        raise FileNotFoundError("shap_values_sample.npz not found — run SHAP cell first.")
    data = np.load(npz, allow_pickle=True)
    shap_values = data['shap_values']

# Use class-1 SHAP values if available (binary)
if isinstance(shap_values, (list, tuple)) and len(shap_values) >= 2:
    shap_vals_for_plot = shap_values[1]
else:
    shap_vals_for_plot = shap_values

# Ensure shap array has shape (n_samples, n_features_trans)
shap_vals_for_plot = np.asarray(shap_vals_for_plot)
if shap_vals_for_plot.ndim == 1:
    # unlikely, but reshape
    shap_vals_for_plot = shap_vals_for_plot.reshape(-1, n_cols_trans)

# final check
if shap_vals_for_plot.shape[1] != n_cols_trans:
    # we still have mismatch — try to reduce to common min columns
    minc = min(shap_vals_for_plot.shape[1], n_cols_trans)
    shap_vals_for_plot = shap_vals_for_plot[:, :minc]
    X_shap_trans = X_shap_trans[:, :minc]
    feat_names = feat_names[:minc]

# Convert transformed matrix to DataFrame with correct column names
X_shap_trans_df = pd.DataFrame(X_shap_trans, columns=feat_names)

# --- 4) Choose feature to plot (PAY_0 preferred, else top SHAP feature) ---
candidate = None
# search for PAY_0 in feat_names (case-insensitive)
for fn in feat_names:
    if 'pay_0' in fn.lower() or 'pay0' in fn.lower():
        candidate = fn
        break

if candidate is None:
    # pick top mean-absolute shap feature
    mean_abs = np.mean(np.abs(shap_vals_for_plot), axis=0)
    top_idx = int(np.argmax(mean_abs))
    candidate = feat_names[top_idx]
    print("PAY_0 not found — selected top SHAP feature:", candidate)

# --- 5) Plot dependence safely ---
out_path = os.path.join(VIZ, 'shap_dependence_pay0_fixed.png')
try:
    import shap
    # use feature name (pandas DataFrame supported)
    shap.dependence_plot(candidate, shap_vals_for_plot, X_shap_trans_df, feature_names=feat_names, show=False)
    plt.title(f"SHAP dependence — {candidate}")
    plt.tight_layout()
    plt.savefig(out_path, dpi=300, bbox_inches='tight')
    plt.show()
    plt.close()
    print("Saved SHAP dependence plot to:", out_path)
except Exception as e:
    plt.close()
    # fallback: simple scatter of feature vs shap value
    print("Dependence_plot failed; falling back to scatter. Error:", e)
    # find index of candidate
    idx = feat_names.index(candidate) if candidate in feat_names else 0
    plt.figure(figsize=(7,5))
    plt.scatter(X_shap_trans_df.iloc[:, idx], shap_vals_for_plot[:, idx], alpha=0.6, s=12)
    plt.xlabel(candidate); plt.ylabel("SHAP value")
    plt.title(f"SHAP scatter (fallback) — {candidate}")
    fallback_path = os.path.join(VIZ, 'shap_dependence_pay0_fallback.png')
    plt.tight_layout()
    plt.savefig(fallback_path, dpi=300, bbox_inches='tight')
    plt.show()
    plt.close()
    print("Saved fallback scatter to:", fallback_path)
```

```python
# NEW CELL — Robust SHAP local plots (waterfall) for TP/FP/FN (handles list/array outputs)
import os, numpy as np, matplotlib.pyplot as plt
import shap, pickle

ROOT = '/content/drive/MyDrive/XAI_TRY_1_PROJECT'
VIZ = os.path.join(ROOT, 'viz')
ARTIFACTS = os.path.join(ROOT, 'artifacts')
os.makedirs(VIZ, exist_ok=True)
os.makedirs(ARTIFACTS, exist_ok=True)

# Ensure model exists
model = globals().get('best_rf', globals().get('rf_pipeline', None))
if model is None:
    raise RuntimeError("No RF model found. Run RF cells first.")

rf_clf = model.named_steps['clf']
explainer = shap.TreeExplainer(rf_clf)

# predictions and true labels
y_pred_test = model.predict(X_test)
y_true_test = y_test.to_numpy().ravel()

# find indices
def find_indices():
    tp = fp = fn = None
    for i in range(len(y_true_test)):
        if y_true_test[i] == 1 and y_pred_test[i] == 1 and tp is None:
            tp = i
        if y_true_test[i] == 0 and y_pred_test[i] == 1 and fp is None:
            fp = i
        if y_true_test[i] == 1 and y_pred_test[i] == 0 and fn is None:
            fn = i
        if tp is not None and fp is not None and fn is not None:
            break
    return tp, fp, fn

tp_idx, fp_idx, fn_idx = find_indices()
print("TP:", tp_idx, "FP:", fp_idx, "FN:", fn_idx)

# helper to get single-instance shap vector (class-1 if available)
def get_shap_vector_for_row(X_row_trans):
    """
    X_row_trans: array-like with shape (1, n_features_trans)
    returns: (shap_vec, base_value) where shap_vec shape = (n_features_trans,)
    """
    raw_shap = explainer.shap_values(X_row_trans)
    # Cases:
    # 1) raw_shap is list/tuple of arrays: [class0_array, class1_array] where each is (n_samples, n_features)
    # 2) raw_shap is array shape (n_samples, n_features) -> binary case flattened
    # 3) raw_shap may be (n_classes, n_samples, n_features) -> handle similarly
    if isinstance(raw_shap, (list,tuple)) and len(raw_shap) >= 2:
        # pick class 1
        arr = np.asarray(raw_shap[1])
        shap_vec = arr[0]  # first (and only) sample
        # base value: explainer.expected_value might be scalar or array per class
        base = explainer.expected_value[1] if hasattr(explainer, 'expected_value') and len(np.atleast_1d(explainer.expected_value))>1 else (explainer.expected_value if hasattr(explainer,'expected_value') else None)
        return shap_vec, base
    else:
        arr = np.asarray(raw_shap)
        if arr.ndim == 2:
            # shape (n_samples, n_features)
            return arr[0], (explainer.expected_value if hasattr(explainer,'expected_value') else None)
        if arr.ndim == 3:
            # maybe (n_classes, n_samples, n_features) -> pick class 1
            if arr.shape[0] >= 2:
                return arr[1,0,:], (explainer.expected_value[1] if hasattr(explainer,'expected_value') and len(np.atleast_1d(explainer.expected_value))>1 else None)
            else:
                return arr[0,0,:], (explainer.expected_value if hasattr(explainer,'expected_value') else None)
    # fallback – try to coerce
    raise ValueError("Unexpected shap value shape: " + str(np.asarray(raw_shap).shape))

# build feature names matching transform
def build_feature_names(ct, trans_mat):
    names = []
    for name, transformer, cols in ct.transformers_:
        if name == 'remainder':
            continue
        if name == 'num':
            names.extend(list(cols))
        elif name == 'cat':
            try:
                ohe = transformer.named_steps.get('ohe', None)
                if ohe is not None:
                    names.extend(list(ohe.get_feature_names_out(cols)))
                else:
                    names.extend(list(cols))
            except Exception:
                names.extend(list(cols))
        else:
            try:
                names.extend(list(cols))
            except Exception:
                pass
    if not names:
        # fallback to generic
        return [f"feat_{i}" for i in range(trans_mat.shape[1])]
    # ensure length match
    if len(names) != trans_mat.shape[1]:
        if len(names) > trans_mat.shape[1]:
            names = names[:trans_mat.shape[1]]
        else:
            need = trans_mat.shape[1] - len(names)
            names += [f"feat_extra_{i}" for i in range(need)]
    return names

# function to create waterfall (or fallback bar) for a given relative index
def create_local_plot(rel_idx, label):
    if rel_idx is None:
        print(f"No {label} instance found — skipping.")
        return None
    # get row as DataFrame and transform
    X_row = X_test.iloc[[rel_idx]]
    X_row_trans = preprocessor.transform(X_row)  # shape (1, n_features_trans)
    feat_names = build_feature_names(preprocessor, X_row_trans)
    # get shap vector + base
    try:
        shap_vec, base_val = get_shap_vector_for_row(X_row_trans)
    except Exception as e:
        print(f"Failed to extract shap vector for {label}: {e}")
        shap_vec = None
        base_val = None
    out_png = os.path.join(VIZ, f"shap_force_inst_{label}.png")
    try:
        if shap_vec is None:
            raise RuntimeError("no shap_vec")
        # create shap.Explanation for single sample
        try:
            # shap expects data to be 1D for a single sample
            data_for_exp = X_row_trans[0] if hasattr(X_row_trans, "__array__") else X_row_trans
            expl = shap.Explanation(values=shap_vec, base_values=base_val, data=data_for_exp, feature_names=feat_names)
            shap.plots.waterfall(expl, show=False)
            plt.title(f"SHAP waterfall — {label} (true={y_true_test[rel_idx]})")
            plt.tight_layout()
            plt.savefig(out_png, dpi=300, bbox_inches='tight')
            plt.close()
            print(f"Saved waterfall for {label} -> {out_png}")
            return out_png
        except Exception as e:
            # fallback: use legacy waterfall or bar plot
            print(f"Waterfall plot failed (fallback) for {label}: {e}")
            # create bar of top contributions
            idxs = np.argsort(np.abs(shap_vec))[::-1][:30]
            names = [feat_names[i] for i in idxs]
            vals = shap_vec[idxs]
            plt.figure(figsize=(8,6))
            plt.barh(range(len(names))[::-1], vals[::-1])
            plt.yticks(range(len(names)), names[::-1], fontsize=8)
            plt.title(f"Top SHAP contributions — {label} (true={y_true_test[rel_idx]})")
            plt.tight_layout()
            plt.savefig(out_png, dpi=300, bbox_inches='tight')
            plt.close()
            print(f"Saved fallback bar for {label} -> {out_png}")
            return out_png
    except Exception as e:
        print(f"Final failure for {label}: {e}")
        return None

# create plots
tp_png = create_local_plot(tp_idx, 'TP')
fp_png = create_local_plot(fp_idx, 'FP')
fn_png = create_local_plot(fn_idx, 'FN')

print("Outputs:")
print(" TP:", tp_png)
print(" FP:", fp_png)
print(" FN:", fn_png)
```

```python
# NEW CELL — Robust LIME local explanations (replacement for previous LIME cell)
import os, json, numpy as np, matplotlib.pyplot as plt
import pandas as pd

ROOT = '/content/drive/MyDrive/XAI_TRY_1_PROJECT'
VIZ = os.path.join(ROOT, 'viz')
ARTIFACTS = os.path.join(ROOT, 'artifacts')
os.makedirs(VIZ, exist_ok=True)
os.makedirs(ARTIFACTS, exist_ok=True)

# get model
model = globals().get('best_rf', globals().get('rf_pipeline', None))
if model is None:
    raise RuntimeError("No RF model found. Run RF baseline/tuning cells first.")

# import lime
from lime.lime_tabular import LimeTabularExplainer

# Prepare data for explainer
X_train_tab = X_train.reset_index(drop=True)
y_train_tab = y_train.reset_index(drop=True)
feature_names = list(X_train_tab.columns)

# determine categorical feature indices robustly
try:
    categorical_features = [feature_names.index(c) for c in cat_cols if c in feature_names]
except Exception:
    categorical_features = [i for i,c in enumerate(feature_names) if X_train_tab.dtypes[c] == 'object'] if len(feature_names)>0 else []

explainer = LimeTabularExplainer(
    training_data = X_train_tab.values,
    feature_names = feature_names,
    class_names = ['No Default','Default'],
    categorical_features = categorical_features if len(categorical_features)>0 else None,
    discretize_continuous = True,
    random_state = 42
)

# find TP/FP/FN indices on test
y_pred_test = model.predict(X_test)
y_true_test = y_test.to_numpy().ravel()
tp_idx = fp_idx = fn_idx = None
for i in range(len(y_true_test)):
    if y_true_test[i] == 1 and y_pred_test[i] == 1 and tp_idx is None:
        tp_idx = i
    if y_true_test[i] == 0 and y_pred_test[i] == 1 and fp_idx is None:
        fp_idx = i
    if y_true_test[i] == 1 and y_pred_test[i] == 0 and fn_idx is None:
        fn_idx = i
    if tp_idx is not None and fp_idx is not None and fn_idx is not None:
        break

print("Indices (relative to X_test): TP, FP, FN =", tp_idx, fp_idx, fn_idx)

def safe_extract_lime_list(exp, pred_label):
    """
    Return a list of (feature_str, weight_float) robustly from a lime explanation object `exp`.
    Tries exp.as_list(label), exp.as_map(), and sensible fallbacks.
    """
    # 1) try as_list with label
    try:
        maybe = exp.as_list(label=pred_label)
        # sometimes as_list returns list of tuples -> good
        if isinstance(maybe, list) and all(isinstance(t, tuple) and len(t)==2 for t in maybe):
            return maybe
        # sometimes as_list can return a scalar or other shape -> ignore
    except Exception:
        pass

    # 2) try as_map (gives dict: label -> list of (idx, weight))
    try:
        amap = exp.as_map()
        if isinstance(amap, dict) and pred_label in amap:
            pairs = []
            for idx, weight in amap[pred_label]:
                fname = feature_names[idx] if idx < len(feature_names) else f"feat_{idx}"
                pairs.append((fname, float(weight)))
            if pairs:
                return pairs
    except Exception:
        pass

    # 3) try exp.as_list() without label (sometimes returns list of (label, list) or list of tuples)
    try:
        maybe2 = exp.as_list()
        # if format is [(feat, weight), ...] -> accept
        if isinstance(maybe2, list) and all(isinstance(t, tuple) and len(t)==2 for t in maybe2):
            return maybe2
        # if format is [(label, [(feat, weight), ...]), ...] -> find predicted label entry
        if isinstance(maybe2, list) and len(maybe2)>0 and isinstance(maybe2[0], tuple) and isinstance(maybe2[0][1], list):
            # find entry for pred_label if present
            for item in maybe2:
                lbl, lst = item
                if lbl == pred_label and isinstance(lst, list):
                    return lst
    except Exception:
        pass

    # 4) As last resort, try exp.as_map() keys and flatten first available
    try:
        amap2 = exp.as_map()
        for lbl, arr in amap2.items():
            pairs = []
            for idx, weight in arr:
                fname = feature_names[idx] if idx < len(feature_names) else f"feat_{idx}"
                pairs.append((fname, float(weight)))
            if pairs:
                return pairs
    except Exception:
        pass

    # Final fallback: return top-10 zeros with feature names
    return [(fn, 0.0) for fn in feature_names[:10]]

def make_lime_plot(test_idx, label):
    if test_idx is None:
        print(f"No {label} instance found — skipping.")
        return None
    x_row = X_test.iloc[test_idx].values
    # predict_fn: accept 2D numpy -> return predict_proba
    predict_fn = lambda arr: model.predict_proba(pd.DataFrame(arr, columns=feature_names))
    exp = explainer.explain_instance(data_row=x_row, predict_fn=predict_fn, num_features=20)
    pred_label = int(model.predict(X_test.iloc[[test_idx]])[0])

    local_list = safe_extract_lime_list(exp, pred_label)
    # build dataframe
    rows = [{'feature': feat, 'weight': float(w)} for feat, w in local_list]
    df = pd.DataFrame(rows)

    # save JSON
    json_path = os.path.join(ARTIFACTS, f'lime_inst_{label}.json')
    with open(json_path, 'w') as f:
        json.dump({'index': int(test_idx), 'pred_label': int(pred_label), 'explanation': rows}, f, indent=2)

    # plot horizontal bar
    plt.figure(figsize=(8,6))
    topk = df.shape[0]
    y_labels = df['feature'][:topk].tolist()[::-1]
    weights = df['weight'][:topk].values[::-1]
    plt.barh(range(len(y_labels)), weights, align='center')
    plt.yticks(range(len(y_labels)), y_labels, fontsize=9)
    plt.xlabel('LIME weight (positive => increases probability of predicted class)')
    plt.title(f'LIME local explanation — {label} (index {test_idx}, pred={pred_label})')
    plt.tight_layout()
    out_png = os.path.join(VIZ, f'lime_inst_{label}.png')
    plt.savefig(out_png, dpi=300, bbox_inches='tight')
    plt.close()
    print(f"Saved LIME plot for {label} ->", out_png)
    return out_png, json_path

# run for TP, FP, FN
tp_out = make_lime_plot(tp_idx, 'TP')
fp_out = make_lime_plot(fp_idx, 'FP')
fn_out = make_lime_plot(fn_idx, 'FN')

print("Outputs:")
print(" TP:", tp_out)
print(" FP:", fp_out)
print(" FN:", fn_out)
```

```python
# NEW CELL — Step 5: UMAP / t-SNE of TEST set colored by RF predicted probability
# Saves: viz/umap_probs.png (or viz/tsne_probs.png fallback) and artifacts/umap_embedding.csv

import os, numpy as np, pandas as pd, matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

ROOT = '/content/drive/MyDrive/XAI_TRY_1_PROJECT'
VIZ = os.path.join(ROOT, 'viz')
ARTIFACTS = os.path.join(ROOT, 'artifacts')
os.makedirs(VIZ, exist_ok=True)
os.makedirs(ARTIFACTS, exist_ok=True)

# Model and data checks
model = globals().get('best_rf', globals().get('rf_pipeline', None))
if model is None:
    raise RuntimeError("No trained model found (best_rf or rf_pipeline). Run RF cells first.")

# raw X_test DataFrame and predicted probability of positive class
X_test_df = X_test.copy().reset_index(drop=True)
if hasattr(model, "predict_proba"):
    probs = model.predict_proba(X_test_df)[:, 1]
else:
    # fallback: use decision_function normalized
    try:
        from sklearn.preprocessing import MinMaxScaler
        scores = model.decision_function(X_test_df)
        probs = MinMaxScaler().fit_transform(scores.reshape(-1,1)).ravel()
    except Exception:
        # last resort: use predicted labels
        probs = model.predict(X_test_df)

# Subsample for embedding (max 5000)
MAX_POINTS = 5000
n = len(X_test_df)
if n > MAX_POINTS:
    rng = np.random.default_rng(42)
    idx = rng.choice(n, MAX_POINTS, replace=False)
    X_sub = X_test_df.iloc[idx].reset_index(drop=True)
    prob_sub = probs[idx]
else:
    idx = np.arange(n)
    X_sub = X_test_df.reset_index(drop=True)
    prob_sub = probs

# Preprocess numeric-only embedding input:
# If there are categorical columns, convert to numeric using preprocessor's numeric columns or simple encoding.
try:
    # attempt to use the fitted preprocessor to get a dense numeric matrix
    X_emb_in = model.named_steps['preproc'].transform(X_sub)
    # if sparse, convert to dense
    try:
        X_emb_in = X_emb_in.toarray()
    except Exception:
        pass
except Exception:
    # fallback: take numeric columns only and scale
    num_cols = X_sub.select_dtypes(include=[np.number]).columns.tolist()
    if not num_cols:
        # fallback: try to coerce all columns to numeric by factorizing categoricals
        X_tmp = X_sub.copy()
        for c in X_tmp.columns:
            if X_tmp[c].dtype == object:
                X_tmp[c] = pd.factorize(X_tmp[c])[0]
        num_cols = X_tmp.columns.tolist()
        X_emb_in = X_tmp.values
    else:
        X_emb_in = X_sub[num_cols].values
    # scale
    X_emb_in = StandardScaler().fit_transform(X_emb_in)

# Try UMAP first, else t-SNE
use_umap = False
try:
    import umap
    use_umap = True
except Exception:
    try:
        import umap.umap_ as umap  # alternate import name in some installs
        use_umap = True
    except Exception:
        use_umap = False

if use_umap:
    try:
        reducer = umap.UMAP(n_components=2, random_state=42, n_neighbors=30, min_dist=0.1)
        emb = reducer.fit_transform(X_emb_in)
        out_png = os.path.join(VIZ, 'umap_probs.png')
        print("Used UMAP for embedding.")
    except Exception as e:
        print("UMAP failed, falling back to t-SNE. Error:", e)
        use_umap = False

if not use_umap:
    from sklearn.manifold import TSNE
    reducer = TSNE(n_components=2, random_state=42, perplexity=30, n_iter=1000, init='pca')
    emb = reducer.fit_transform(X_emb_in)
    out_png = os.path.join(VIZ, 'tsne_probs.png')
    print("Used t-SNE for embedding.")

# Build DataFrame with embedding and probs (and optional original index)
emb_df = pd.DataFrame(emb, columns=['x','y'])
emb_df['prob_default'] = prob_sub
emb_df['orig_index'] = idx
csv_out = os.path.join(ARTIFACTS, 'umap_embedding.csv')
emb_df.to_csv(csv_out, index=False)
print("Saved embedding CSV to:", csv_out)

# Plot scatter colored by predicted probability
plt.figure(figsize=(8,6))
sc = plt.scatter(emb_df['x'], emb_df['y'], c=emb_df['prob_default'], cmap='viridis', s=12, alpha=0.9)
plt.colorbar(sc, label='Predicted prob(default)')
plt.title('UMAP / t-SNE of TEST set — colored by predicted default probability')
plt.xlabel('dim1'); plt.ylabel('dim2')
plt.tight_layout()
plt.savefig(out_png, dpi=300, bbox_inches='tight')
plt.show()
plt.close()
print("Saved embedding plot to:", out_png)
```

```python
# NEW CELL — Step 6: Calibration plot (reliability curve) for Random Forest
import os, json, numpy as np, matplotlib.pyplot as plt
from sklearn.calibration import CalibrationDisplay, calibration_curve
from sklearn.metrics import brier_score_loss

ROOT = '/content/drive/MyDrive/XAI_TRY_1_PROJECT'
VIZ = os.path.join(ROOT, 'viz')
ARTIFACTS = os.path.join(ROOT, 'artifacts')
os.makedirs(VIZ, exist_ok=True)
os.makedirs(ARTIFACTS, exist_ok=True)

# choose model
model = globals().get('best_rf', globals().get('rf_pipeline', None))
if model is None:
    raise RuntimeError("No RF model pipeline found (best_rf or rf_pipeline). Run RF cells first.")

# ensure X_test and y_test exist
X_eval = X_test.copy()
y_eval = y_test.to_numpy().ravel()

# predicted probabilities
if hasattr(model, "predict_proba"):
    y_prob = model.predict_proba(X_eval)[:,1]
else:
    # fallback to decision_function then minmax scale
    try:
        scores = model.decision_function(X_eval)
        from sklearn.preprocessing import MinMaxScaler
        y_prob = MinMaxScaler().fit_transform(scores.reshape(-1,1)).ravel()
    except Exception:
        # last fallback: predicted labels as 0/1
        y_prob = model.predict(X_eval)

# Brier score
brier = float(brier_score_loss(y_eval, y_prob))

# Calibration curve (reliability)
prob_true, prob_pred = calibration_curve(y_eval, y_prob, n_bins=10, strategy='uniform')

# Plot: reliability curve + histogram
fig, ax = plt.subplots(2,1, figsize=(6,8), gridspec_kw={'height_ratios':[3,1]})
# reliability
ax0 = ax[0]
ax0.plot([0,1],[0,1],'--', color='gray')
ax0.plot(prob_pred, prob_true, marker='o', label=f'RF (Brier={brier:.4f})')
ax0.set_ylabel('Fraction of positives')
ax0.set_xlabel('')
ax0.set_title('Calibration (Reliability) Curve — Random Forest')
ax0.legend(loc='lower right')
ax0.grid(alpha=0.25)

# histogram of probabilities
ax1 = ax[1]
ax1.hist(y_prob, bins=20, range=(0,1), edgecolor='k', alpha=0.6)
ax1.set_xlabel('Predicted probability')
ax1.set_ylabel('Count')
ax1.set_title('Predicted probability distribution (test set)')

plt.tight_layout()
out_png = os.path.join(VIZ, 'calibration_rf.png')
fig.savefig(out_png, dpi=300, bbox_inches='tight')
plt.close(fig)
print("Saved calibration plot to:", out_png)

# Save calibration metrics for reproducibility
metrics = {
    "brier_score": brier,
    "calibration_curve": {"prob_true": prob_true.tolist(), "prob_pred": prob_pred.tolist()},
    "n_bins": 10,
    "n_samples_eval": int(len(y_eval))
}
with open(os.path.join(ARTIFACTS, 'calibration_metrics.json'), 'w') as f:
    json.dump(metrics, f, indent=2)
print("Saved calibration metrics to artifacts/calibration_metrics.json")
```
